Define Map

```python

m = folium.Map(location=[55, -2], zoom_start=5)
```

Define Style

```python
def style_function(feature):
    radius = feature['properties']['ml'] * 5
    return {'radius': radius,
            'weight': 2,
            'fill_opacity': 0}
```

Draw map using API data

```python
# Using requests to access data
folium.GeoJson(
    response.json(), style_function=style_function,
    marker=folium.CircleMarker()).add_to(m)
m
```

```python
# Using OWSLib data
bgs = Features('https://ogcapi.bgs.ac.uk')
# Fails due to F5 issues internally
earthquakes = bgs.collection_items('recentearthquakes', limit=10)

m2 = folium.Map(location=[55, -2], zoom_start=6)

def style_function(feature):
    radius = feature['properties']['ml'] * 5
    return {'radius': radius,
            'weight': 2,
            'fill_opacity': 5}

folium.GeoJson(
    earthquakes, style_function=style_function,
    marker=folium.CircleMarker()).add_to(m2)
m
```
